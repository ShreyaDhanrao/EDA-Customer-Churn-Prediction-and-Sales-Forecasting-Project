```python
df_rem_col['Date'] = pd.to_datetime(df['Date'], format='%m-%d-%y')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
max_value =df_rem_col['Amount'].max()
min_value =df_rem_col['Amount'].min()
print(max_value)
print(min_value)
```

```python
plt.figure(figsize=(12, 10))
sns.barplot(x='Amount', y='Category', data=df,estimator=sum,errorbar=None,palette='muted',hue='Category')
plt.title('Sales by Product Category')
plt.show()
```

```python
plt.figure(figsize=(10, 16))
sns.barplot(x='Amount', y='ship-state', data=df, estimator=sum,errorbar=None,palette='deep',hue='ship-state',legend=False)
plt.title('Sales by Product State')
plt.show()
```

```python
plt.figure(figsize=(12, 30))
sns.barplot(x='Amount', y='Date', data=df, estimator=sum,errorbar=None)
plt.title('Sales by Date')
xticks=plt.gca().get_xticks()
print(xticks)
for i in range(len(xticks)-1):
    lower=xticks[i]
    upper=xticks[i+1]
    print(f"Range:{lower} to {upper}")
plt.show()
```

```python
df_rem_col['Year'] = df_rem_col['Date'].dt.year
```

```python
df_rem_col['Month'] =df_rem_col['Date'].dt.month
```

```python
df_rem_col['Year-Month'] = df_rem_col['Date'].dt.to_period('M')
```

```python
sales_by_month = df_rem_col.groupby('Year-Month')['Amount'].sum()
print(sales_by_month)
```

```python
print(type(sales_by_month))
```

```python
sales_by_month = sales_by_month.to_frame()
```

```python
print(sales_by_month.columns)
```

```python
print(sales_by_month)
```

```python
sales_by_month['Year-Month'] = sales_by_month.index.astype(str)
```

```python
plt.figure(figsize=(14, 7))
sns.lineplot(data=sales_by_month, x='Year-Month', y='Amount')
plt.title('Sales Trends Over Time')
plt.xlabel('Time (Year-Month)')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
sales_by_product = df.groupby('Category')['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False).head(10)

plt.figure(figsize=(14, 7))
sns.barplot(data=sales_by_product, x='Amount', y='Category', palette='viridis',hue='Category',legend=False)
plt.title('Top 10 Products by Sales')
plt.xlabel('Total Sales')
plt.ylabel('Product')
plt.show()
```
